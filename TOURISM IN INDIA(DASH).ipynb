# TOURISM IN INDIA

### OBJECTIVE

● FORIGN EXCHANGE OVER THE YEARS (PERCENTAGE)

● VISITORS OF 2017 2018 ACCORDING TO STATES/UT

● TOP 10 PLACES VISITED IN INDIA (DOMESTIC)

● STATE/UT TRAVELER DATA AND PERCENTAGE

● TO ANALYIZE AND VIZUALIZE THE VISITOR COUNT


### DATA ANALYSIS

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objs as go
import pandas as pd


app = dash.Dash()


df = pd.read_csv('223.csv')
dff = pd.read_csv('2.csv')
dfff = pd.read_csv('111.csv')
dffff = pd.read_csv('1 (3).csv')
dfffff = pd.read_csv('1 (4).csv')
dffffff = pd.read_csv('333.csv')


app.layout = html.Div([html.Div(html.H1('TOURISM IN INDIA',style={'color':'white',
                                                    'text-align':'center',
                                                    'background-color':'black',
                                                    'border-style':'double',
                                                    'display':'inline-block',
                                                    'width':'100%'
                                                    })),
    dcc.Graph(
        id='places',
        style={'width':'98%','item-align':'flex','display':'inline-block'},
        figure={
            'data': [
                go.Bar(
                    x = df['YEARS'],
                    y = df['CHANGE OVER YEARS (rupees)'],
                    
                )
                
            ],
            'layout': go.Layout(
                title = 'FOREIGN EXCHANGE OVER THE YEAR IN RUPEES PERCENTAGE',
                xaxis = {'title': 'YEARS'},
                yaxis = {'title': 'CHANGE OVER YEARS (rupees)'},
                paper_bgcolor = '#F9E79F'
             )
        }
    ),
    
    dcc.Graph(
        id='place1',
        style={'width':'48%','display':'inline-block'},
        figure={
            'data': [
                go.Scatter(
                    x = df['YEARS'],
                    y = df['CHANGE OVER YEARS (dollars)'],
                    mode = 'markers'
                )
                
            ],
            'layout': go.Layout(
                title = 'FOREIGN EXCHANGE EARNING OVER YEARS IN DOLLARS PERCENTAGE',
                xaxis = {'title': 'YEARS'},
                yaxis = {'title': 'CHANGE OVER YEARS (dollars)'},
                paper_bgcolor = '#A2B9CE'
             )
        }
    ),
    
    dcc.Graph(
        id='place2',
        style={'width':'48%','display':'inline-block'},
        figure={
            'data': [
                go.Scatter(
                    x = dff['STATE/UT'],
                    y = dff['SHARE FOREIGN'],
                    name = 'SHARE FOREIGN'
                ),
                
                go.Bar(
                    x = dff['STATE/UT'],
                    y = dff['SHARE DOMESTIC'],
                    name = 'SHARE DOMESTIC'
                )
                
            ],
            'layout': go.Layout(
                title = 'SHARE OF FOREIGN AND DOMESTIC %',
                xaxis = {'title': 'State And Union territories'},
                yaxis = {'title': 'Shares of Foreign And Domestic'},
                paper_bgcolor = '#E5E8E8'
             )
        }
    ),
    
     dcc.Graph(
        id='place3',
        style={'width':'48%','display':'inline-block'},
        figure={
            'data': [
                go.Bar(
                    x = dfff['STATE/UT'],
                    y = dfff['DOMESTIC 2017'],
                    name = 'DOMESTIC 2017'
                ),
                
                go.Bar(
                    x = dfff['STATE/UT'],
                    y = dfff['FOREIGN 2017'],
                    name = 'FOREIGN 2017'
                )
                
            ],
            'layout': go.Layout(
                title = "VISITORS OF 2017 ACCORDING TO STATE/UT",
                xaxis = {'title': 'State And Union territories'},
                yaxis = {'title': 'Vistors Of 2017'},
                barmode="stack",paper_bgcolor = '#D2B4DE'
             )
        }
    ),
    
     dcc.Graph(
        id='place4',
        style={'width':'48%','display':'inline-block'},
        figure={
            'data': [
                go.Bar(
                    x = dfff['STATE/UT'],
                    y = dfff['DOMESTIC 2018'],
                    name = 'DOMESTIC 2018'
                ),
                
                go.Bar(
                    x = dfff['STATE/UT'],
                    y = dfff['FOREIGN 2018'],
                    name = 'FOREIGN 2018'
                )
                
            ],
            'layout': go.Layout(
                title = "VISITORS OF 2018 ACCORDING TO STATE/UT",
                xaxis = {'title': 'State And Union territories'},
                yaxis = {'title': 'Vistors Of 2018'},
                paper_bgcolor = '#F5B7B1'
             )
        }
    ),
    
    
     dcc.Graph(
        id='place5',
         style={'width':'48%','display':'inline-block'},
        figure={
            'data': [
                go.Scatter(
                    x = dffff['Name of Monument'],
                    y = dffff['No. of Domestic visitors'],
                    mode = 'markers+lines',
                )
                
            ],
            'layout': go.Layout(
                title = 'NUMBER OF DOMESTIC VISITORS',
                xaxis = {'title': 'Name of Monument'},
                yaxis = {'title': 'Domestic visitors'},
                paper_bgcolor = '#F6DDCC'
             )
        }
    ),
    
    
    dcc.Graph(
        id='place6',
        style={'width':'48%','display':'inline-block'},
        figure={
            'data': [
                go.Box(
                    x = dffff['Name of Monument'],
                    y = dffff['%'],
                )
                
            ],
            'layout': go.Layout(
                title = 'NUMBER OF DOMESTIC VISITORS IN %',
                xaxis = {'title': 'Name of Monument'},
                yaxis = {'title': 'Domestic visitors in %'},
                barmode = 'stack',paper_bgcolor = '#F7DC6F'
             )
        }
    ),
    
    dcc.Graph(
        id='place7',
        style={'width':'48%','display':'inline-block'},
        figure={
            'data': [
                go.Bar(
                    x = dfffff['Name of Monument'],
                    y = dfffff['No. of Foreign visitors'],
                )
                
            ],
            'layout': go.Layout(
                title = 'TOP 10 VISITED(FOREIGN) PLACES VISITORS NUMBERS',
                xaxis = {'title': 'Name of Monument'},
                yaxis = {'title': 'Foreign visitors'},
                paper_bgcolor = '#FFD700'
             )
        }
    ),
                       
     dcc.Graph(
        id='place8',
        style={'width':'48%','display':'inline-block'},
        figure={
            'data': [
                go.Heatmap(
                    x = dffffff['Name of Countries '],
                    y = dffffff['Yr_2008'],
                    z = dffffff['Yr_2008'],
                    name = '2008'
                ),
                
                 go.Heatmap(
                    x = dffffff['Name of Countries '],
                    y = dffffff['Yr_2009'],
                    z = dffffff['Yr_2008'],
                     name = '2009'
                ), 
                
                go.Heatmap(
                    x = dffffff['Name of Countries '],
                    y = dffffff['Yr_2010'],
                    z = dffffff['Yr_2008'],
                    name = '2010'
                ),
                
                go.Heatmap(
                    x = dffffff['Name of Countries '],
                    y = dffffff['Yr_2011'],
                    z = dffffff['Yr_2008'],
                    name = '2011'
                ),
                
                 go.Heatmap(
                    x = dffffff['Name of Countries '],
                    y = dffffff['Yr_2012'],
                    z = dffffff['Yr_2008'],
                     name = '2012'
                ),
                
                 go.Heatmap(
                    x = dffffff['Name of Countries '],
                    y = dffffff['Yr_2013'],
                    z = dffffff['Yr_2008'],
                     name = '2013'
                ),
                
                 go.Heatmap(
                    x = dffffff['Name of Countries '],
                    y = dffffff['Yr_2014'],
                    z = dffffff['Yr_2008'],
                     name = '2014'
                ),
                
                 go.Heatmap(
                    x = dffffff['Name of Countries '],
                    y = dffffff['Yr_2015'],
                    z = dffffff['Yr_2008'],
                    name = '2015'
                ),
                
            ],
            'layout': go.Layout(
                title = 'FOREIGNS VISITED TO INDIA LAST FEW YEARS',
                xaxis = {'title': 'Name of Countries'},
                yaxis = {'title': 'Vistiors per Years'},
                paper_bgcolor = '#98FB98'
             )
        }
    ),
                       
                       
])
if __name__ == '__main__':
    app.run_server(port=2514)
```
